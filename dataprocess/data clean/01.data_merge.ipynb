```python
# # # 读取学生答题记录
# submit_record = pd.read_csv(relative_path + 'Data_SubmitRecord/SubmitRecord-Class1.csv')

def data_merge(file, submit_record, title_info, student_info):
    error_row_index = [] # 记录错误信息的行索引

    # 获取submit_record的行数
    submit_record_len = submit_record.shape[0]

    for index, row in submit_record.iterrows():
        if index >= submit_record_len:
            break

        student_id = row['student_ID'] # 取出学生ID
        title_id = row['title_ID'] # 取出题目ID

        # 取出学生行信息
        student_info_row = student_info[student_info['student_ID'] == student_id]
        # 如果学生行信息为空，跳过
        if student_info_row.empty:
            print('student_id: ' + student_id + ' not found!')
            error_row_index.append(row.name)
            continue

        # 将取出的学生信息添加到对应的列中
        submit_record.loc[row.name, 'sex'] = student_info_row['sex'].values[0] # 取学生性别
        submit_record.loc[row.name, 'age'] = student_info_row['age'].values[0] # 取学生年龄
        submit_record.loc[row.name, 'major'] = student_info_row['major'].values[0] # 取学生专业

        # 取出对应的题目行信息
        title_info_rows = title_info[title_info['title_ID'] == title_id] 
        # 如果题目行信息为空
        if title_info_rows.empty:
            print('title_id: ' + title_id + ' not found!')
            # 记录此条错误信息的行索引
            error_row_index.append(row.name)
            continue
        
        # 将取出的知识点和满分添加到对应的列中
        # submit_record.loc[row.name, 'knowledge'] = title_info_rows['knowledge'].values[0]
        # submit_record.loc[row.name, 'sub_knowledge'] = title_info_rows['sub_knowledge'].values[0] 
        # submit_record.loc[row.name, 'full_score'] = title_info_rows['score'].values[0]# 取题目对应的满分
        
        # 遍历未处理的title_info_rows 一个题目ID对应可能多个知识点
        first_line = True
        for _, title_info_row in title_info_rows.iterrows(): # _为数据行索引index，不一定从0开始的
            if first_line:
                submit_record.loc[row.name, 'knowledge'] = title_info_row['knowledge']
                submit_record.loc[row.name, 'sub_knowledge'] = title_info_row['sub_knowledge']
                submit_record.loc[row.name, 'full_score'] = title_info_row['score']# 取题目对应的满分
                first_line = False
                continue

            # 在submit_record中末尾添加一行
            dynaminc_len = submit_record.shape[0]
            submit_record.loc[dynaminc_len] = submit_record.iloc[index] # 复制一行数据
            submit_record.loc[dynaminc_len, 'knowledge'] = title_info_row['knowledge']
            submit_record.loc[dynaminc_len, 'sub_knowledge'] = title_info_row['sub_knowledge']
            submit_record.loc[dynaminc_len, 'full_score'] = title_info_row['score']  

    # 删除错误信息的行 inplace=True表示在原数据上删除
    submit_record.drop(error_row_index, inplace=True)

    # 将补齐后的submit_record保存到Merged_Data_SubmitRecord文件夹中
    submit_record.to_csv(relative_path + 'Merged_Data_SubmitRecord/' + file, index=False)
```

```python
# 读取Data_SubmitRecord文件夹中的所有csv文件
for file in os.listdir(relative_path + 'Data_SubmitRecord'):
    print(file + ' merge start!')
    submit_record =  pd.read_csv(relative_path + 'Data_SubmitRecord/' + file)
    data_merge(file, submit_record, title_info, student_info)
    print(file + ' merge success!')
    # break
```
